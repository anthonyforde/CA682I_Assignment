## Create Analytics Base Table

```
base_data = ""
base_data = health_data[['YearStart', 'LocationAbbr', 'LocationDesc', 'Question', 'Data_Value', 'StratificationID1', 'StratificationID2']].copy()
base_data
```

#### Select required fields from BRFSS ageing mental health questionnaire

```
base_data.columns = ['Year', 'State_Abbr', 'State', 'Question', 'Perc_Mental_Distress', 'Age_Category', 'Gender']
base_data
```

#### Filter on required question - the subject of this research: 
"Percentage of older adults who are experiencing frequent mental distress"

```
base_data = base_data[base_data['Question'] == "Percentage of older adults who are experiencing frequent mental distress"] 
base_data
```

#### From a stratification breakdown, for this study, we are only concerned with gender breakdown: male, female and overall. Filter out ethnic breakdown which is outside the scope of this research.

```
base_data = base_data[(base_data['Gender'] == "MALE") | (base_data['Gender'] == "FEMALE") | (base_data['Gender'] == "OVERALL")] 
base_data
```

#### Filter out US territories and regions - we are only concerned with the 50 US states, DC and US overall data.

```
base_data = base_data[base_data.State_Abbr != "GU"]   # GUAM
base_data = base_data[base_data.State_Abbr != "MDW"]  # Midwest
base_data = base_data[base_data.State_Abbr != "NRE"]  # Northeast
base_data = base_data[base_data.State_Abbr != "PR"]   # Puerto Rico
base_data = base_data[base_data.State_Abbr != "SOU"]  # South
base_data = base_data[base_data.State_Abbr != "VI"]   # Virgin
base_data = base_data[base_data.State_Abbr != "WEST"] # West
```

```
base_data['Age_Category'] = base_data['Age_Category'].astype(str)
```

```
base_data.loc[base_data["Age_Category"] == "AGE_OVERALL", "Age_Category"] = "All"
base_data.loc[base_data["Age_Category"] == "65PLUS", "Age_Category"] = ">65"
base_data.loc[base_data["Age_Category"] == "5064", "Age_Category"] = "<65"
base_data.loc[base_data["Gender"] == "OVERALL", "Gender"] = "All"
base_data.loc[base_data["Gender"] == "MALE", "Gender"] = "Male"
base_data.loc[base_data["Gender"] == "FEMALE", "Gender"] = "Female"
```

```
base_data
```

```
base_data.isna().sum()
```

Deal with null values

```
base_data[base_data['Perc_Mental_Distress'].isna()]
```

```
base_data.loc[base_data["Perc_Mental_Distress"].isna(), "Perc_Mental_Distress"] = 0
```

#### Append new columns to base data in preparation for other data sources

```
base_data['Average_Medicare_Cost'] = 0
base_data['Average_Premium'] = 0
base_data['US_Population'] = 0
base_data['US_Population_Over_65'] = 0
base_data['US_Perc_Over_65'] = 0
base_data['State_Population'] = 0
base_data['State_Population_Over_65'] = 0
base_data['State_Perc_Over_65'] = 0
```

```
base_data.loc[base_data["State"] == "United States, DC & Territories", "State"] = "United States"
```

```
for index, row in population_age_profile.iterrows():
    base_data.loc[(base_data.Year == row['Year']) & (base_data.State_Abbr == row['State_Code']), 'US_Population'] = row['US_Population']
    base_data.loc[(base_data.Year == row['Year']) & (base_data.State_Abbr == row['State_Code']), 'US_Population_Over_65'] = row['US_Population_Over_65']
    base_data.loc[(base_data.Year == row['Year']) & (base_data.State_Abbr == row['State_Code']), 'US_Perc_Over_65'] = row['US_Perc_Over_65']
    base_data.loc[(base_data.Year == row['Year']) & (base_data.State_Abbr == row['State_Code']), 'State_Population'] = row['State_Population']
    base_data.loc[(base_data.Year == row['Year']) & (base_data.State_Abbr == row['State_Code']), 'State_Population_Over_65'] = row['State_Population_Over_65']
    base_data.loc[(base_data.Year == row['Year']) & (base_data.State_Abbr == row['State_Code']), 'State_Perc_Over_65'] = row['State_Perc_Over_65']

base_data
```
